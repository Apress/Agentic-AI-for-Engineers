```python
"""
Single-Agent Technical Article Generator (CrewAI)
------------------------------------------------

This example shows the simplest possible CrewAI setup:
1. Define a single agent (Senior Technical Writer)
2. Define a single task (write an article)
3. Pass a topic at runtime
4. Generate a polished article using an LLM behind the agent

This version includes:
âœ” Robust, structured prompts
âœ” Beginner-friendly comments
âœ” Clear responsibilities and guardrails
âœ” Safety and formatting rules
"""

# ================================
# ðŸ”§ IMPORTS & MODEL SETUP
# ================================
from crewai import Agent, Task, Crew, LLM
import os
from dotenv import load_dotenv

load_dotenv()  # Load environment variables from .env


# ---------------------------- MODEL SETUP ----------------------------
# CrewAI wrapper around an OpenAI model
llm = LLM(model="gpt-4o-mini")
 
 

# ================================
# ðŸ¤– DEFINE THE AGENT
# ================================
senior_technical_writer = Agent(
    role="Senior Technical Writer",
    
    # Clear mission the agent tries to fulfill
    goal=(
        "Create clear, engaging, accurate technical articles "
        "that simplify complex concepts while remaining detailed and structured."
    ),
    
    # This narrative helps shape writing style
    backstory=(
        "You are an experienced technical writer known for transforming "
        "complex engineering topics into easy-to-read, professional documentation. "
        "You structure information logically, avoid jargon overload, and ensure "
        "technical accuracy in all explanations."
    ),
    
    llm=llm,
    verbose=True
)


# ================================
# ðŸ“ DEFINE THE TASK
# ================================
writing_task = Task(
    description="""
Write a **well-structured, engaging, and technically accurate article** on the topic:
{topic}

Your article MUST follow this outline:

# 1. Introduction  
- Briefly explain the topic  
- Why it matters  
- Who benefits  

# 2. Core Concepts  
- Define key terms  
- Explain components, architecture, or mechanics  
- Use simple examples  

# 3. How It Works  
- Step-by-step breakdown  
- Real-world workflow  
- Practical explanation  

# 4. Applications & Use Cases  
- At least 3 real or realistic use cases  
- Why they matter  

# 5. Challenges & Limitations  
- Be honest and technically accurate  
- No fear-mongering or hype  

# 6. Future Outlook  
- What developments are expected  
- Practical predictions only  

# 7. Conclusion  
- Summarize the article  
- Reinforce relevance  
- Keep positive and realistic

=========================
IMPORTANT WRITING RULES
=========================

- DO NOT hallucinate nonexistent research or facts.  
- If unsure about something, say so transparently.  
- Write in simple, readable paragraphs.  
- Prefer clarity over complexity.  
- Avoid marketing tone or excessive hype.  
- Output **only the final article**, no extra commentary.

""",

    expected_output="A polished, structured, and accurate technical article about the given topic.",
    agent=senior_technical_writer
)


# ================================
# ðŸ§  BUILD THE CREW
# ================================
crew = Crew(
    agents=[senior_technical_writer],
    tasks=[writing_task],
    verbose=True
)


# ================================
# ðŸš€ RUN THE WORKFLOW
# ================================
response = crew.kickoff(inputs={"topic": "AI Agents"})


# ================================
# ðŸ–¨ SHOW OUTPUT
# ================================
print("\n\n================ FINAL ARTICLE OUTPUT ================\n")
print(response)
```
